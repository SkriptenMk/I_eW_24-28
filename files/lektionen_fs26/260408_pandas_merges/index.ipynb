### Context

You already know how to combine tables using SQL `JOIN`. Today you learn
how to do the same thing in Python with **pandas** — the tool you'll use
whenever your data lives in CSV files, spreadsheets, or API responses
rather than in a database.

### Key Idea

| SQL | pandas |
|-----|--------|
| `INNER JOIN` | `df1.merge(df2, on='key', how='inner')` |
| `LEFT JOIN` | `df1.merge(df2, on='key', how='left')` |
| `RIGHT JOIN` | `df1.merge(df2, on='key', how='right')` |
| `ON t1.col = t2.col` | `on='col'` |
| `NULL` | `NaN` |

### Three-Way Merge

**Tables:** `class_roster` + `test_scores` (see Task 1) +

| Table | Columns |
|---|---|
| `homework` | `student_id` (int), `homework_avg` (float) |

Mr. Weber now also has a `homework` table. Merge all three tables
(`class_roster`, `test_scores`, `homework`) and compute a final average
grade for each student.

Hint: You can chain merges: `a.merge(b, ...).merge(c, ...)`

````python
# Stretch goal data
homework = pd.DataFrame({
    'student_id': [1, 2, 3, 4, 5, 6],
    'homework_avg': [5.0, 4.5, 5.5, 4.0, 5.5, 3.5]
})

# Your code here:
````

<details>
<summary>Click to reveal the solution</summary>

```python
combined = class_roster.merge(test_scores, on='student_id', how='left') \
                       .merge(homework, on='student_id', how='left')
combined['final_avg'] = combined[['score', 'homework_avg']].mean(axis=1)
print(combined[['name', 'score', 'homework_avg', 'final_avg']])
```

Students without a test score get their homework average as the final grade (since `mean` ignores `NaN` by default).
